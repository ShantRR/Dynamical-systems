# Dynamical Systems Overview (Reminder)

**Formal definition**

A dynamical system is a 3-tuple $(T,S,\Phi)$

$S$ is a set of dynamical system states, $T$ is a set of evolution parameter/time values and $\Phi$ is the evolution function of dynamical system.
$$\Phi: D\subseteq T\times S \to S$$

$$\Phi(0,x) = x\quad\text{initial condition}$$

$$\Phi(t_2,\Phi(t_1, x)) = \Phi(t_1+t_2,x)$$


**Flows & Maps / continous and discrete times**

$\bullet$ If $T\subseteq \mathbb{R}$, $\Phi$ continous and differentiable function than we have differential euqations, flows.

$$\dot{x}=f(x)$$

$\bullet$ If $T\subseteq \mathbb{Z}$, $\Phi$ function than we have maps.

$$x_{n+1} = f(x_n)$$

**Connection between maps and flows**

$\bullet$ discretization of differential equations for numerical computing (Euler method, Runge-Kutta methods, e.t.c)

$$\dot{x}=f(x)\quad \text{Euler method}\quad x_{n+1}=x_n+hf(x_n)$$

$\bullet$ Poincar√© section of flow

Maps also used for
$\bullet$ Discretization of algebraic equations 

**Conservative and Dissipative systems**

$\bullet$ If $\text{div}(f(x)) = 0$ (for flows) or $\det(J(f)) = 1$ (for maps) called conservative dynamical systems.

$(M, \mu, \varphi_{t})$\
Monifold $M$\
measure $\mu$ in $M$ $\qquad \mu(M)=1$\
and $\varphi_{t}: M\rightarrow M $ is a measure preserving transformation, i.e. $\mu(A) = \mu (\varphi_t\,A )$ 

$\bullet$ If $\text{div}(f(x)) < 0$ (for flows) or $\det(J(f)) < 1$ (for maps) called dissipative dynamical systems.




# Period doubling

## 1-dimensional maps Overview

1 dimensional iterative maps
$$x_{n+1}=f(x_n),\quad x_n\in\mathbb{R}$$
generate the orbit
$\{x_0,f(x_0),f^2(x_0),\dots, f^n(x_0)\}$

$$f^n(x):=(f\circ f\circ \dots \circ f)\,(x)$$

**Fixed points** of $f$: $$f(x_i^*)=x_i^*$$

if $x^*$ is a fixed point of $f$ it is also fixed point of $f^n,\,\, \forall n\in\mathbb{N}$, i.e. $$f^n(x^*)=x^*$$

it can be illustrated graphically


**Periodic orbits/cycles** of $f$: 

$$\{x_1^*,x_2^*,\dots,x_p^*| x_i^*\neq x_j^*, i\neq j\}=:\mathcal{P}$$

\begin{equation}
\begin{cases}
x_{i+1}^*=f(x_{i}^*),\quad i = 1,2,\dots,p-1 \\
x_{1}^*=f(x_p^*)
\end{cases}
\end{equation}


or equivalently

$$f^p(x_i^*)=x_i^*$$

*For example:* 2-cycle:

\begin{equation}
\begin{cases}
f(x_1^*)=x_2^*\\
f(x_2^*)=x_1^*
\end{cases}
\equiv
\begin{cases}
f^2(x_1^*)=x_1^*\\
f^2(x_2^*)=x_2^*
\end{cases}
\end{equation}

**Stability:**

*Fixed points*

Let's $x^*$ is a fixed point of the function $f$

$$x^*+\tilde{x}_n$$

$$\tilde{x}_{n+1}+x^* = f(x^*+\tilde{x}_n)\approx  f(x^*)+\frac{d f(x)}{dx}\Big|_{x^*}\tilde{x}_n+\frac{d^2 f(x)}{dx^2}\Big|_{x^*}\tilde{x}_n^2+\dots$$

$$\tilde{x}_{n+1}=\mu \tilde{x}_n \quad \text{(multiplicator)} \mu:=f'(x^*)\quad \Rightarrow$$ 

$\bullet$ if $|f'(x^*)|>1$ unstable

$\bullet$ if $|f'(x^*)|<1$ stable

$\bullet$ if $|f'(x^*)|=0$ superstable


*Periodic orbits*

$$(f^{n}(x))'=f'(x) f'(f(x)) \dots f'(f^{n-1}(x))$$

$\forall x_i^*,x_j^*\in\mathcal{P}: \,\, (f^p(x_i^*))'=(f^p(x_j^*))'$

*For example:* 2-cycle:

$(f^2(x))'|_{x_1^*}=f'(x)|_{x_1^*}f'(f(x))|_{x_1^*}=f'(x_1^*)f'(x_2^*)$\
$(f^2(x))'|_{x_2^*}=f'(x)|_{x_2^*}f'(f(x))|_{x_2^*}=f'(x_2^*)f'(x_1^*)$


**Bifurcations:**

Let's consider the following map $$x_{n+1}=f(x_n;\lambda)$$

Whith changing a parameter $\lambda$, the behaviour of the dynamics also changes

for a period doubling bifurcation realization in 1-dimensional maps it is necessary:

$$(Sf)(x):=\frac{f'''(x)}{f'(x)}-\frac{3}{2}\left(\frac{f''(x)}{f'(x)}\right)^2$$

$$(Sf)(x)<0$$

$\square$ Let $f(x)$ and $x^*$: $f'(x^*)=-1,\quad f^2(x):=f(f(x))\quad \Rightarrow$

\begin{align}
(f^2(x))'|_{x^*}&=+1\qquad ((f^2(x))'|_{x^*}=f'(f(x))|_{x^*}f'(x)|_{x^*}=f'(x)|_{x^*}f'(x)|_{x^*}=(-1)^2)\\
(f^2(x))''|_{x^*}&=0\\
(f^2(x))'''|_{x^*} &= -2f'''(x)-3[f''(x)]^2
\end{align}

and for the schwarzian derivative in the point $x^*$ we have

$$(Sf)(x) = -f'''(x)-\frac{3}{2}[f''(x)]^2\quad \Rightarrow$$

$$(f^2(x))'''|_{x^*} = 2(Sf)(x)$$

what happens from the point of view $f^2(x)$ when $x^*$ loses stability 

$$x_n = x^*+\tilde{x}_n \quad\Rightarrow$$

$$\tilde{x}_{n+1} = \tilde{x}_n + \frac{1}{6}(f^2(x))'''|_{x^*}\tilde{x}_n^3\qquad\text{(pitchfork bifurcation)}\qquad\blacksquare$$

**Lyapunov Exponent, Chaotic behaviour**

Let $x_0$ is the initil point and considering another initial point $x_0+\delta_0$

$$|\delta_n|=|\delta_0|e^{n\sigma}$$

\begin{align}
\sigma &= \lim_{\delta_0\to 0 \\ n\to\infty}\left\{\frac{1}{n}\ln\Bigg|\frac{\delta_n}{\delta_0}\Bigg|\right\}\\
&= \lim_{\delta_0\to 0\\ n\to\infty}\left\{\frac{1}{n}\ln\Bigg|\frac{f^n(x_0+\delta_0)-f^n(x_0)}{\delta_0}\Bigg|\right\}\\
&=\lim_{n\to\infty}\left\{\frac{1}{n}\ln|(f^n(x_0))'|\right\}=\lim_{n\to\infty}\left\{\frac{1}{n}\sum_{i=0}^{n-1}\ln|f'(x_i)|\right\}
\end{align}

## N-dimensional maps

$$\mathbf{x}_{n+1}=\mathbf{f}(\mathbf{x}_n),\quad \mathbf{x}_n\in\mathbb{R}^{N}$$

**Fixed point of** $\mathbf{f}$: Similarly as in 1-dimensional case $$\mathbf{f}(\mathbf{x}_i^*)=\mathbf{x}_i^*$$
Also the cycles is defined in a similar way to the one-dimensional maps. For example 2-dimensional map

\begin{align}
x_{n+1}&=f(x_n,y_n)\\
y_{n+1}&=g(x_n,y_n)
\end{align}
period 2 cycle must be satisfy

\begin{align}
x_2^*=f(x_1^*,y_1^*)&\quad x_1^*=f(x_2^*,y_2^*)\\
y_2^*=g(x_1^*,y_1^*)&\quad y_1^*=g(x_2^*,y_2^*)
\end{align}

**Stability**:

*fixed points*

Jacobian of function $\mathbf{f}(\mathbf{x})$
$$\mathbb{D}f\equiv \mathbf{J}(\mathbf{f})$$

$$\mathbf{J}_{ij}:=\frac{\partial f_i(\mathbf{x})}{\partial x_j}$$


$$\tilde{\mathbf{x}}_{n+1}=\mathbf{J}\tilde{\mathbf{x}}_n$$

the stability of fixed points are determinated by the eigenvalues $\det (J-\lambda_i \mathbb{1})=0$ of Jacobian.

### 2-dimension case

$\bullet\quad \lambda_{1,2}\in\mathbb{R}$ and $\lambda_1\neq\lambda_2$

$$J\xrightarrow{\substack{\exists\,\,\text{Linear} \\ \text{transformation}}} \begin{pmatrix}
\lambda_1 & 0 \\
0 & \lambda_2
\end{pmatrix}$$

stable iff $|\lambda_{1}|<1$ and $|\lambda_2|<1$

$\bullet\quad \lambda_{1}=\lambda_{2}\equiv\lambda\in\mathbb{R}$

$$J\xrightarrow{\substack{\exists\,\, \text{Linear} \\ \text{transformation}}} \begin{pmatrix}
\lambda & 1 \\
0 & \lambda
\end{pmatrix}$$

stable iff $|\lambda|<1$


$\bullet\quad \lambda_{1,2}=\alpha\pm i\beta$

$$J\xrightarrow{\substack{\exists\,\,\text{Linear} \\ \text{transformation}}} \begin{pmatrix}
\alpha & -\beta \\
\beta & \alpha
\end{pmatrix}$$

stable iff $|\lambda|=\alpha^2+\beta^2<1$

**Stability triangle**

$$\lambda_{1,2}= \frac{\text{Tr}(J)}{2}\pm\sqrt{\frac{\text{Tr}(J)^2}{4}-\det(J)}$$
